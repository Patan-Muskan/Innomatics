```python
def smart_parking(capacity, logs):
    parked = 0
    peak_usage = 0

    # process logs
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        # track peak usage
        if parked > peak_usage:
            peak_usage = parked

    # status check
    if parked > capacity:
        status = "ALERT: Parking Full / Exceeded!"
    else:
        status = "Available"

    # output
    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak_usage)
    print("Parking Status:", status)


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, vehicle_logs)
```

```python
def estimate_delivery_time(distance, traffic, weather):
    # base speed (km per hour)
    speed = 30

    # base time in minutes
    base_time = (distance / speed) * 60

    # traffic delay
    if traffic == "High":
        base_time += 20
    elif traffic == "Medium":
        base_time += 10
    elif traffic == "Low":
        base_time += 5

    # weather delay
    if weather == "Rainy":
        base_time += 15
    elif weather == "Stormy":
        base_time += 25
    elif weather == "Clear":
        base_time += 0

    # final output
    print("Estimated Delivery Time:", int(base_time), "minutes")


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic, weather)
```

```python
def analyze_seat_booking(total_seats, booked_seats):
    # count booked seats
    booked_count = len(booked_seats)

    # calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    # determine show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 70:
        status = "Almost Full"
    else:
        status = "Seats Available"

    # suggestion for extra shows
    if occupancy >= 90:
        suggestion = "Suggestion: Open additional shows"
    else:
        suggestion = "No additional show required"

    # output
    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)
    print(suggestion)


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 booked entries

analyze_seat_booking(total_seats, booked_seats)
```

```python
def classify_server_load(cpu_readings):
    total = 0

    # loop to calculate sum
    for value in cpu_readings:
        total += value

    # average calculation
    average = total / len(cpu_readings)

    # classification
    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    # output
    print("Average CPU Load:", int(average), "%")
    print("Server Status:", status)


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

```python
def resource_tracker(usage):
    OVERUSE_LIMIT = 8
    overused = []

    # check resource usage
    for resource, hours in usage.items():
        if hours > OVERUSE_LIMIT:
            overused.append(resource)

    # output
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_tracker(usage_data)
```

```python
def manage_registrations(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0

    # process registrations
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # status check
    if confirmed == capacity:
        status = "Closed"
    else:
        status = "Open"

    # output
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
capacity = 100
registrations = 105

manage_registrations(capacity, registrations)
```
